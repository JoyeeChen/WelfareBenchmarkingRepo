Ideal standard for below: use relative file paths using "./logs...", not absolute ones, for efficient refactoring later. 

How to easily tell what log (out of the possibly many logs in the log folder) to use? 
Note their timestamps. The timestamps used on .eval-format folders use UTC time zone, while the timestamps used on wandb documentation like the "Run data is saved locally in" printouts, for code blocks above, use your local time zone.

```python
rrs( #copy and paste relative eval log path in!
    logs = "./logs/2025-12-16T02-58-03+00-00_ahb_DFFvKPTopef5v2Bg9xUYdw.eval",
    seed = 21,
    n_elems = 15,
)
```

The outputs of rrs method can be manipulated too!

```python
output = rrs(
    logs = "./logs/2025-12-16T02-58-03+00-00_ahb_DFFvKPTopef5v2Bg9xUYdw.eval",
    seed = 21,
    n_elems = 15,
)
```
